```python
#Import our libraries
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import gzip
import sys
import warnings

from ctgan import CTGANSynthesizer
```

```python
#Import our data
data = pd.read_csv('/content/drive/MyDrive/artificial_genome_project/test_dataset.csv')
data = data.iloc[:, 2:]
data.head()
```

```python
from ctgan import CTGANSynthesizer
ctgan = CTGANSynthesizer(epochs=100)
#specify columns that are going to be used for training (ie. get rid of ID, pair)
discrete_columns = data.columns.tolist()
```

```python
print(discrete_columns)
```

```python
ctgan.fit(data, discrete_columns)

#Make synthetic samples from the data.
samples = ctgan.sample(1000)
```

```python
#save model
ctgan.save('/content/drive/MyDrive/artificial_genome_project/baseline_model.pkl')
```

```python
#save synthesized data
samples.to_csv('/content/drive/MyDrive/artificial_genome_project/baseline_synthesized_dataset.csv', index=False)
```
